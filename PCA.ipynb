```python
## standardizing the data
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
wine_arr=wine.values
```

```python
x=sc.fit_transform(wine_arr)
x
# x is the standardized data
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA()
```

```python
wine_pca=pca.fit_transform(x)
wine_pca
```

```python
pca.explained_variance_ratio_
```

```python
var=np.cumsum(pca.explained_variance_ratio_*100)
```

```python
np.round(var,3)
```

```python
plt.plot(var)
```

```python
final=pd.concat([df['Type'],pd.DataFrame(wine_pca[:,0:3],columns=['PC1','PC2','PC3'])],axis=1)
final
```

```python
# visualizing pca1,pca2,pca3
plt.figure(figsize=(18,12))
sns.scatterplot(data=final)
```

# Heirarchial clustering

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hc=AgglomerativeClustering(n_clusters=3)
```

```python
# Plotting dendrograms
from scipy.cluster import hierarchy
```

```python
hierarchy.linkage(x,method='ward')
dend=hierarchy.dendrogram(hierarchy.linkage(x,method='complete'))
```

```python
ypred=hc.fit_predict(x)
```

```python
df['cluster HC']=ypred
df
```

# KMeans clustering

```python
from sklearn.cluster import KMeans
km=KMeans()
```

```python
wcss=[]

for i in range(1,6):
    km=KMeans(n_clusters=i)
    km.fit_predict(x)
    wcss.append(km.inertia_)
    
plt.plot(range(1,6),wcss)    
```

```python
# here figure clearly shows that 3 is the optimum number of cluster
km=KMeans(n_clusters=3)
```

```python
y=km.fit_predict(x)
```

```python
df['cluster KM']=y
df
```
